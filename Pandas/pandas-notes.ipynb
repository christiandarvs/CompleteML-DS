### Manipulate Data with N/A

```python
missing_data = pd.read_csv('car-sales-missing-data-clone.csv')
display(missing_data)
```

### Fill Empty Cells

```python
missing_data['Odometer'] = missing_data['Odometer'].fillna(40_000)
missing_data
```

```python
# However, it's important to note that using inplace=True can be risky if you're not careful, because it modifies the original DataFrame and it can't be undone. Therefore, it's recommended to make a backup of your data before using inplace=True
```

### Drop data that are N/A

```python
dropped_data = missing_data.dropna()
dropped_data
```

### Create a new feature and INSERT in the DataFrame

```python
seats = pd.Series([5, 4, 4, 7, 4])
missing_data['Seats'] = seats
missing_data
```

### Fill N/A cells in Seats Column

```python
missing_data['Seats'] = missing_data['Seats'].fillna(5)
missing_data
```

### Create a new feature using Python List


```python
# From Python List
fuel_capacity = [42, 32, 26, 42, 42, 26, 42, 62, 52, 42]
missing_data['Fuel Capacity'] = fuel_capacity
missing_data
# If we are going to use a list, the length needs to match the length of the dataset or else it will return an ERROR
```

### Create a new feature using a formula

```python
fuel_per_100km = pd.Series([7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3, 4.5])
missing_data['Fuel Per 100 KM'] = fuel_per_100km
missing_data
```

```python
missing_data['Total Fuel Used'] = missing_data['Odometer'] / 100 * missing_data['Fuel Per 100 KM']
missing_data
```

### Create a new feature by assigning a fixed value

```python
missing_data['Number of Wheels'] = 4
missing_data
```

```python
missing_data['Passed Road Safety'] = True
missing_data
```

### Drop a COLUMN

```python
missing_data.drop('Passed Road Safety', axis=1, inplace=True)
missing_data
```

```python
missing_data.drop('Doors', axis=1, inplace=True)
missing_data
```

### Shuffle the order of the index

```python
shuffled_data = missing_data.sample(frac=1)
# 1 means 100% of the dataset, 0.5 means 50% of the dataset
# The order of the dataset will be shuffled, but the information doesn't
shuffled_data
```

### Reset the order of index

```python
shuffled_data.reset_index(drop=True, inplace=True)
shuffled_data
```

### Apply a lamba function to a specific column

```python
missing_data['Odometer'] = missing_data['Odometer'].apply(lambda x: x * 5)
missing_data
```
