### Add Data

```python
x = [1, 2, 3, 4]
y = [10, 30, 20, 50]
plt.plot(x, y);

# they need to have the same length of elements
```

### 1st Method

```python
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

### 2nd Method

```python
fig = plt.figure() # create a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y) # add some data
plt.show()
```

### 3rd Method (Recommended)

```python
fig, ax = plt.subplots()
# Create a figure and a set of subplots.
# This utility wrapper makes it convenient to create common layouts of subplots, including the enclosing figure object, in a single call.
plt.title('Recommended Approach')
ax.plot(x, y);
```

### Which approach is recommended according to ChatGPT:
This method combines the creation of both the figure and the axes in a single line. It's a convenient way to create a single plot. If you don't need to specify the layout of multiple subplots, this method is often the most concise and recommended approach. - 3rd Method

In practice, the third method (plt.subplots()) is often favored for creating single plots due to its simplicity and readability. If you need to create multiple subplots in a single figure, plt.subplots() can also handle this easily.

The second method (add_axes) is useful when you need precise control over the position and size of the axes, which can be important in complex layouts.

Ultimately, the choice of method depends on your specific use case and personal preference.

### Matplotlib Example Workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5))

# 3. Plot data
ax.plot(x, y);

# 4. Customize Plot
ax.set(title='Sample Plot', xlabel='x-axis', ylabel='y-axis');
fig.savefig(fname='../Matplotlib - Plotting and Data Visualization/sample-plot.png')
```
